# Read in data and call package

```r

LD <- read.csv("~/Desktop/Data Sci for Psych & Neuro/LexicalData_toclean.csv")

head(LD)

library(tidyverse)
```

# Remove commas and change D_RT to numeric/check var class

```r


LD[,'D_RT'] <- gsub(",","",LD[,'D_RT'])

View(LD)
LD$D_RT<-as.numeric(LD$D_RT)
is.numeric(LD$D_RT)

```

# Remove RT rows with missing values and make sure all RT are positive

```r

cleaned_LD <- LD  %>% 
    filter(LD != 'NA',
           D_RT >= 0)
           
head(cleaned_LD)
```

# Addressing data anomalies

# Summary stats in table including number of obs., mean RT, and SD of RT.

```r

 smry <- cleaned_LD  %>% 
    summarise(n_observations = length(Sub_ID), # how many observations observations are there total? 
              mean_RT = mean(D_RT), #mean reaction times
              RT_sd = sd(D_RT), #SD reaction times
              )
smry
```
