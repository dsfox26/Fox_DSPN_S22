## Spatially related play at age 4 predicts preschoolers’ spatial skills at age 5

### Background

Spatial cognition is a unique component of general intelligence as well as a predictor of math performance (LeFevre et al., 2010; Mix & Cheng, 2012). The term spatial cognition refers to a range of cognitive abilities related to visualizing, manipulating, and transforming the objects and spaces we encounter daily (Uttal et al., 2013). Some of these abilities are present in infancy and undergo a protracted development, extending to include more complex and abstract competencies over time. For example, one study has shown that infants with greater spatial abilities measured via their ability to discriminate between objects rotated in space vs mirrored objects are better at mentally transforming shapes at age 4 (Lauer & Lourenco, 2016). Moreover, children’s spatial and patterning abilities can uniquely predict later math skills (Rittle-Johnson, Zippert, & Boice, 2019), and greater math ability is linked in turn to better health and medical decision-making (Reyna et al., 2009). In addition, greater spatial cognition in childhood is linked to greater educational and occupational outcomes, especially in STEM fields (Wai, Lubinski, & Benbow, 2009). Thus, spatial cognition acts as a foundational skill that impacts human capital outcomes throughout life, however the origins of variability in spatial cognition are still unclear. 


Previous studies have shown that engaging in spatially related activities and play (e.g., puzzles and block building) positively influences spatial thinking and reasoning in children and adults (Baenninger & Newcombe 1989; Baenninger & Newcombe, 1995; Casey et al., 2008; Cherney, 2008; Costa-Giomi, 1999; Doyle et al., 2012; Ozel et al., 2004; Weckbacher & Okamoto, 2012). For example, Levine and colleagues (2012) observed naturalistic play in 2- to 4-year-old children and their parents over a two-year span and found that those who played with puzzles outperformed those who did not on a mental transformation task. Results from a study conducted by Jirout and Newcombe (2015) confirm that more frequent spatial play (i.e., puzzles, blocks, and board games) as reported by the parents is associated with stronger spatial skills in 4 -7-year-olds. 
Further longitudinal work showed that participation in spatial activities in childhood predicted spatially related problem-solving strategies as well as participation in spatial activities in adolescence (Peterson et al., 2020). In sum, engagement in spatial activities is associated with better spatial skills throughout development. 

It is the aim of the present study to investigate whether parents' reports of their children's spatial activity within the home at age 4 is predictive of their spatial skills at age 5. Children (N=113) and their parents participated in a one-year longitudinal study starting when children were four years old. Measures of children’s mental transformation and sensitivity to spatial features were administered at each time point. At time 1 (T1) parents reported over two phone interviews whether or not their children engaged in four specific spatial activities (block, puzzle, shape, and matching play) at home.

### Methods
##### Specify all variables, their data type, how they were collected, and number of observations.

**Participants:**

Data were drawn from a longitudinal study of 113 socioeconomically diverse parent-child dyads. Participating children were 51% female and, on average, 4 years, 4 months (SD = 0.3, range = 4.0 - 4.9) at the first time point (T1). Time 2 (T2) data were collected one year later when children were 5 years of age. 

**Procedures:** 

All data for T1 were collected in person over the span of two home visits and two phone calls. Due to the COVID-19 pandemic, all data for T2 were collected via videoconferencing. Families without the required technology were loaned a laptop with a built-in webcam and a mobile Wi-Fi hotspot.


### Measures 

**Children's spatial skills at T1 & T2:**

At both time points children's spatial skills were assessed using the Children’s Mental Transformation Task (CMTT)and the Geometric Sensitivity (GS) test. Spatial skills at T1 and T2 were determined by creating a composite score using children's results for these two measures. 

The CMTT is a measure of children’s mental transformation skills, specifically two-dimensional mental transformations, including horizontal and diagonal translations and rotations (Levine, et al. 1999). Participants were presented with two shape pieces and asked to identify the shape that those pieces would create if they were put together from a set of four response options (see Figure 1).


The GS test (Dehaene et al., 2006) was developed to measure children’s ability to perceive differences in spatial relations and geometric properties of 2D visual displays. Each display features six different images, five of which share a geometric property that is not present in the sixth image (see Figure 2). Children are asked to point to the picture that does not belong. Displays include geometric properties like distance, length, angles, symmetry, topology, straight lines, and parallelism.



**Children's daily spatial activity at T1:**
Over two phone interviews parents were asked a series of questions regarding whether or not their child engaged in specific spatially related activities (block, puzzle, shape, and matching play). A composite was created by averaging the number of activities parents reported their children participating in. 

Previous research using survey measures of parent-reported spatial activities usually asked about frequencies of activities over the span of one to four weeks using questionnaires with Likert scale response options like “never”, “rarely” (once or twice per week), “sometimes” (three to five times per week), and “often” (six or more times per week; Siegel-Hinson & McKeever, 2002; Jirout & Newcombe, 2015; Oostermeijer et al., 2014; Newcombe et al., 1983), or required parents to make a mark on a continuum line between “never” and “always” (Doyle et al., 2012). While these reports may average over day-to-day fluctuations in children’s activities, this method of reporting may introduce recall bias, memory errors, and inflated frequencies to appear more educationally oriented (Bachman et al., 2020). The present study seeks to reduce reporting bias by using parent reports from the previous day examining the number of spatially relevant activities a child engaged in over the past 24 hours. 


**Child age at T1**
Age was measured by calculating the number of days from the participant’s date of birth to the date of the first home visit, then dividing by 30 to determine children’s age in months.

### Research Questions & Hypotheses 

Explicitly state your hypothesis in terms of the structure of the models that you will be evaluating (e.g., What is the form of Y = f(X) for each hypothesis? Will you be using linear regression or classifier models?)

RQ1: Does the frequency of children's spatial activities at age 4 predict preschoolers’ spatial skills at age 5?


H1: It is the hypothesis of the present study that frequency of children’s daily spatial activities will positively predict spatial skills at age 5 such that participating in more frequent spatial activities at age 4 will correspond with greater spatial skills.

_H_1) $ Y_{Spatial\,SkillsT2} = \hat{\beta}_0 + \hat{\beta}_{Spatial\,Activity} + \hat{\beta}_{Spatial\,Skills\,T1} + \hat{\beta}_{Child\,Age}$

Alternatively, there may be no association between children's spatial activities at age 4 and their spatial skills at age 5, and children's spatial skills at age 5 are only predicted by their spatial skill at age 4.


_H_0) $ Y_{Spatial\,SkillsT2} = \hat{\beta}_0 + \hat{\beta}_{Spatial\,Skills\,T1}$

### General Linear Model for Inference

This analysis tested whether there is a relationship between parents' reports of spatially relevant activities at age 4 and children's spatial skills at age 5 (controlling for age and spatial skills at age 4). Standardized beta coefficients were also calculated to determine the strength of the association.

```r
#Linear model addressing RQ:  Does frequency of children's spatial activities at age 4 predict preschoolers’ spatial skills at age 5?

lm.beta
function (MOD) 
{
  b <- summary(MOD)$coef[-1, 1]
  sx <- sd(MOD$model[-1])
  sy <- sd(MOD$model[1])
  beta <- b * sx/sy
  return(beta)
}

modA <- lm(st.c2spatial ~ st.d1aspat + st.x1cage + st.c1spatial, data = dsa)
summary(modA)
coef(modA)
confint(modA)
lm.beta(modA)
```

```r
modno <- lm(st.c2spatial ~ st.c1spatial, data = dsa)
summary(modno)
```

```r
#Plotting Model 1

sp_talk_long%>%ggplot(aes(x=d1aspat, y=c2spatial)) +
  geom_point()+
  labs(x= "Daily Spatial Activities T1", y="Spatial Skills T2")+
  geom_smooth(method= "lm", formula = "y~x", se=TRUE)+
  geom_jitter()+
  theme_minimal()
```

#### Summarizing Regression Results

Our regression results indicate that there is a relationship between children's daily spatial activities at age 4 and their spatial skills at age 5 even while controlling for spatial skills at age 4, such that children who engaged in more spatial activities had better spatial skills one year later. Our a priori alpha value of p<.05 was met, and the adjusted R^2 value indicates that our model accounts for 32% of the variance found in children's spatial skills at T2. Additionally, the confidence intervals for the daily spatial activity variable do not include 0, indicating that we can reject the null hypothesis.
I also found the standardized beta coefficients for the IVs in order to examine the strength of each of their effects on children's spatial skills at T2. We can see that a one unit increase in daily spatial activities is associated with a 21% increase in spatial skills. 

I also ran a regression looking at the null hypothesis that there is not relation between children's spatial activities and their spatial skills, however the model that includes children's spatial activities at age 4 has a higher R^2 value than the null, indicating that more variance in children's spatial skills at T2 is accounted for when including their spatial activities in the model. This will be further tested by calculating Bayes Factors.

### Bootstrapping

```r
library(boot)

boot.fn <- function(data, index){  
    # return: throw this as output
    # coef: extract coefficients from model object 
    return(coef(lm(st.c2spatial ~ st.d1aspat + st.x1cage + st.c1spatial, data=data, subset=index)))}
print(boot.fn(dsa, 1:93)) 


print(coef(lm(st.c2spatial ~ st.d1aspat + st.x1cage + st.c1spatial, data=dsa)))


##the coefficients are the same, sanity check passed
```

```r
boot_obj = boot(dsa ,boot.fn ,R=1000) #R=repetitions 
print(boot_obj)

#t1=intercept, t2= daily spatial activities, t3= child age, t4= child spatial skills at T1
```

```r
attributes(boot_obj)
```

```r
hist(boot(dsa ,boot.fn ,R=1000)$t[,2], xlab="Daily Sp Act Coeff") 
                           
```

#### Bootstrapping Summary

Through bootstrapping, we randomly resampled the existing data with replacement to determine what range of coefficient values we get for the daily spatial activity variable. We can see that the most frequent coefficient value is around .10. Our bootstrapped coefficients are the same as the coefficients resulting from the regression model.

### Bayes Factor

```r
invisible(library(BayesFactor))
```

```r
#bayesfactor for activities
full<-lmBF(st.c2spatial ~ st.d1aspat + st.x1cage + st.c1spatial, data=dsa)
null<-lmBF(st.c2spatial ~ st.c1spatial , data=dsa)

baes <- c(full, null)
baes

full/null
```

```r
plot(baes)
```

#### Bayes Factor Summary 

The Bayes Factors for H1 and H0 are very similar, and after directly comparing them the ratio between the two is 1.15 which means there is not strong evidence that the H1 model better explains the variance in children's spatial skills at age 5 than the H0. 

### Conclusions

The aim of the present study was to determine if children’s daily spatial activity frequency is predictive of their growth in spatial ability from age 4 to age 5. We tested this hypothesis using several methods of data anyalysis. First, regression analyses indicated that the frequency of spatially relevant activities at age 4 was predictive of children’s spatial ability one year later even while controlling for age and spatial ability at age 4 (R2 = .32, F (3,89) =15.51, p<.05). 

Bootstrap testing revealed similar results to the regression anaylsis which added confidence to rejecting the null hypothesis that children's spatially relevant activities are predictive of their spatial skills.

However, calculating Bayes factor for H1 and H0 revealed that there is not strong evidence to support our H1. Though, prior literature exists that suggests support for our H1.

In conclusion, there is some evidence that children's spatial activities at age 4 as reported by parents are predictive of their spatial skills one year later. While the current results extend our knowledge of which environmental factors contribute to children’s spatial skill development, a number of limitations need to be noted. One limitation is that the interview only included questions about four activities: puzzle play, building play, shape sorting, and matching. There are additional spatially relevant activities that young children regularly engage in like board games, spatial video games, and sports (Cherney, 2008; Doyle, 2012; Ho et al., 2018; Lee et al., 2019; Ozel et al., 2004). In the future, it would be beneficial to include a broader range of spatial activity questions and evaluate which activities are more highly correlated to spatial skills than others.

The current study was also limited by the measures of children’s spatial skills employed. Both measures were forced choice tasks that did not tap into spatial language comprehension or production. Future research should include other response formats and measures that tap into other aspects of children’s spatial cognition (e.g., spatial language, 3D mental rotation of 3D objects, 2D and 3D match to sample assembly tasks).
